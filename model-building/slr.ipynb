# Modelling simple linear regression

```python
import pandas as pd
df = pd.read_csv('../data/dataset_ALL.csv')
```

```python
df.head()
```

```python
X = df['living_area'].values
y = df['price'].values
print(type(X))
print(type(y))
print(X.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x = df['living_area'], y = df['price'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 42)
```

```python
from sklearn.linear_model import LinearRegression

X_train = X_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_train,y_train)
```

```python
X_test = X_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
regressor.score(X_test, y_test)
```

```python
y_pred = regressor.predict(X_test)
print(y_pred)
```

```python
plt.plot(X_train, regressor.predict(X_train))
plt.scatter(X_train,y_train, color = 'red')
plt.title('living_area vs price (Training set)')
plt.xlabel('living_area')
plt.ylabel('price')
plt.show()
```

```python
plt.plot(X_test, regressor.predict(X_test))
plt.scatter(X_test,y_test, color = 'red')
plt.title('living_area vs price (Testing set)')
plt.xlabel('living_area')
plt.ylabel('price')
plt.show()
```
